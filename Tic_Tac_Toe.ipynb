```python
import random   #This library is used to generate random numbers
```

```python
def display_board(value):
    print("\n")
    print("\t     |     |")
    print("\t  {}  |  {}  |  {}".format(value[0], value[1], value[2]))
    print('\t_____|_____|_____')
 
    print("\t     |     |")
    print("\t  {}  |  {}  |  {}".format(value[3], value[4], value[5]))
    print('\t_____|_____|_____')
 
    print("\t     |     |")
 
    print("\t  {}  |  {}  |  {}".format(value[6], value[7], value[8]))
    print("\t     |     |")
    print("\n")
```

```python
#This fuction displays scoreboard
def display_scoreboard(score_board):
    print("\t--------------------------------")
    print("\t              SCOREBOARD       ")
    print("\t--------------------------------")
 
    player = list(score_board.keys())
    print("\t   ", player[0], "\t    ", score_board[player[0]])
    print("\t   ", player[1], "\t    ", score_board[player[1]])
 
    print("\t--------------------------------\n")
```

```python
#This function checks if any player has won the game
def check_win(player_position, current_player):
 
    #These are all possible winning combinations
    win_combination = [[1, 2, 3], [4, 5, 6], [7, 8, 9], [1, 4, 7], [2, 5, 8], [3, 6, 9], [1, 5, 9], [3, 5, 7]]
 
    #This is used to check if any combination is matched or not.
    for x in win_combination:
        if all(y in player_position[current_player] for y in x):
          return True #If any combination is matched
       
    return False #If no combination is matched 
```

```python
#This function checks whether game is draw or not.
def check_draws(player_position):
    if len(player_position['X']) + len(player_position['O']) == 9:
        return True
    return False      
```

```python
#This function is used for single game
def single(current_player):
 
    
    value = [' ' for x in range(9)] #Empty list has been created to represent the board.
    player_position = {'X':[], 'O':[]} #This dictionary stores the places occupied by X and O.
     
  
    while True:
        display_board(value)
        print("Player ",current_player," your turn now type mark to select automatic mode or type man to select manual mode : ", end="")
        cho=input()
        if cho == "man" or cho == "Man" or cho == "MAN" or cho == "manual" or cho == "Manual" or cho == "MANUAL":
              #Here we have used Try-exception to check if the value is an integer and between 1 to 9.
          try:
            move = int(input("Please enter a empty position between 1 to 9  : ")) 
          except ValueError:
            print("Wrong Input!!! Please Try Again")
            continue
        elif cho == "mark" or cho == "Mark" or cho == "MARK":
          try:
            move = random.randint(1, 9)
            print(current_player," you have got ",move)
          except ValueError:
            print("Wrong Input!!! Please Try Again")
            continue
        else:
          print("Wrong Input!!! Please Try Again")
          continue
 
        #Checking if move>1 and move<9
        if move < 1 or move > 9:
            print("Wrong Input!!! Please Try Again")
            continue
 
        #Checking if the place is already filled or not.
        if value[move-1] != ' ':
            print("Place already filled. Please Try again!!")
            continue
 
        
 
        #Here we are updating the empty list which was created. 
        value[move-1] = current_player
 
        #Here we are updating the player's position
        player_position[current_player].append(move)
 
        #Here checking if the player has won the game or not.
        if check_win(player_position, current_player):
            display_board(value)
            print("Congratulations!!! Player ", current_player, " has won the game!!")     
            print("\n")
            return current_player
 
        #Here checking if game ends in a draw or not.
        if check_draws(player_position):
            display_board(value)
            print("Game ends in a Draw")
            print("\n")
            return 'D'
 
        #Here switching the player move 
        if current_player == 'X':
            current_player = 'O'
        else:
            current_player = 'X'
```

```python
if __name__ == "__main__":
    print("\n***************WELCOME TO Tic Tac Toe GAME***************\n")
    print("Player 1")
    player1 = input("Enter the name : ")
    print("\n")
 
    print("Player 2")
    player2 = input("Enter the name : ")
    print("\n")
     
    #Here we store the player which chooses between X and O
    current_player = player1
 
    #Here we store the choices of player
    player_choice = {'X' : "", 'O' : ""}
 
    #Here we store the option
    option = ['X', 'O']
 
    #This stores the scoreboard which is intially 0.
    score_board = {player1: 0, player2: 0}
    display_scoreboard(score_board)
 
    #This loop runs multiple times till the time player want to quit
    while True:
 
        
        print("Turn to choose for", current_player)
        print("Enter 1 for X")
        print("Enter 2 for O")
        print("Enter 3 to Quit")
 
        # Try exception for choices input
        try:
            choices = int(input())   
        except ValueError:
            print("Wrong Input!!! Please Try Again\n")
            continue
 
        # Condition for choice of player  
        if choices == 1:
            player_choice['X'] = current_player
            if current_player == player1:
                player_choice['O'] = player2
            else:
                player_choice['O'] = player1
 
        elif choices == 2:
            player_choice['O'] = current_player
            if current_player == player1:
                player_choice['X'] = player2
            else:
                player_choice['X'] = player1
         
        elif choices == 3:
            print("Final Scores")
            display_scoreboard(score_board)
            break  
 
        else:
            print("Wrong Choice!!!! Please Try Again\n")
 
        # This stores the winner in a single game
        winner = single(option[choices-1])
         
        # This edits scoreboard according to result
        if winner != 'D' :
            player_won = player_choice[winner]
            score_board[player_won] = score_board[player_won] + 1
 
        display_scoreboard(score_board)
        # Switch player who chooses X or O means if the players want to continue to play again then next time player 2 gets a chance to choose between X and O
        if current_player == player1:
            current_player = player2
        else:
            current_player = player1
```
